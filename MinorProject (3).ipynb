```python
import pandas as pd
import matplotlib.pyplot as py
```

```python
pf=pd.read_csv('StudentsPerformance (1) (1).csv')
print("->stats of females")
female=pf.loc[pf['gender']=="female"]
print(female.mean())

print('\n->stats of males')
male=pf.loc[pf['gender']=="male"]
print(male.mean())

#bar graph

meanOfMale=male.mean()
meanOfFemale=female.mean()

data = {"meanOfMale":[meanOfMale[i] for i in range(3)],
       "meanOfFemale":[meanOfFemale[0],meanOfFemale[1],meanOfFemale[2]]
       }

index     = ["math", "reading", "writing"]
dataFrame = pd.DataFrame(data=data, index=index)
dataFrame.plot.bar(rot=15, title="male V/S female result comparision")
py.show()
print("In math,males have performed better then females where as in reading and writing females have performed better")
 
    
```

```python
print("->females with standard lunch")
stdf=female.loc[female['lunch']=="standard"]
print(stdf.mean())

print("\n->females with reduced(free) lunch")
freef=female.loc[female['lunch']=="free/reduced"]
print(freef.mean())

#bar graph

meanOfstd=stdf.mean()
meanOffree=freef.mean()

data = {"Females with standard lunch":[meanOfstd[0],meanOfstd[1],meanOfstd[2]],

        "Females with free lunch":[meanOffree[0],meanOffree[1],meanOffree[2]]
         }

index     = ["math", "reading", "writing"]
dataFrame = pd.DataFrame(data=data, index=index)
dataFrame.plot.bar(rot=15, title="comparision of females with different type of lunch")
py.show()
print("Females with standard lunch have performed better")
```

```python
print("->males with standard lunch")
stdm=male.loc[male['lunch']=="standard"]
print(stdm.mean())
print("\n->males with reduced(free) lunch")
freem=male.loc[male['lunch']=="free/reduced"]
print(freem.mean())

#bar graph

meanOfstd=stdm.mean()
meanOffree=freem.mean()

data = {"males with standard lunch":[meanOfstd[0],meanOfstd[1],meanOfstd[2]],

        "males with free lunch":[meanOffree[0],meanOffree[1],meanOffree[2]]
         }

index     = ["math", "reading", "writing"]
dataFrame = pd.DataFrame(data=data, index=index)
dataFrame.plot.bar(rot=15, title="comparision of males with different type of lunch")
py.show()
print("Males with standard lunch have performed better")
```

```python
print("->females who completed the course")
comf=female.loc[female['test preparation course']=="completed"]
print(comf.mean())
print("\n->females who didn't completed the course")
nonef=female.loc[female['test preparation course']=="none"]
print(nonef.mean())

#bar graph

meanOfstd=comf.mean()
meanOffree=nonef.mean()

data = {"females who completed the course":[meanOfstd[0],meanOfstd[1],meanOfstd[2]],

        "females who didn't completed the course":[meanOffree[0],meanOffree[1],meanOffree[2]]
         }

index     = ["math", "reading", "writing"]
dataFrame = pd.DataFrame(data=data, index=index)
dataFrame.plot.bar(rot=15, title="comparision of females on basis of course completion")
py.show()
print("Females who have completed the course have performed better")
```

```python
print("->females of group A")
fa=female.loc[female['race/ethnicity']=="group A"]
print(fa.mean())
print("\n->females of group B")
fb=female.loc[female['race/ethnicity']=="group B"]
print(fb.mean())
print("\n->females of group C")
fc=female.loc[female['race/ethnicity']=="group C"]
print(fc.mean())
print("\n->females of group D")
fd=female.loc[female['race/ethnicity']=="group D"]
print(fd.mean())
print("\n->females of group E")
fe=female.loc[female['race/ethnicity']=="group E"]
print(fe.mean())

#bar graph

meanfA=fa.mean()
meanfB=fb.mean()
meanfC=fc.mean()
meanfD=fd.mean()
meanfE=fe.mean()

data = {"female of group A":[meanfA[0],meanfA[1],meanfA[2]],
        "female of group B":[meanfB[0],meanfB[1],meanfB[2]],
        "female of group C":[meanfC[0],meanfC[1],meanfC[2]],
        "female of group D":[meanfD[0],meanfD[1],meanfD[2]],
        "female of group E":[meanfE[0],meanfE[1],meanfE[2]]
         }

index     = ["math", "reading", "writing"]
dataFrame = pd.DataFrame(data=data, index=index)
dataFrame.plot.bar(rot=15, title="comparision of females of different race")
py.show()
print("performance of group E girls was best where as group A girls have performed worst")
```
